```python
def createfile(name):
   
    date = datetime.now()
    date_dmy = date.strftime("%d%b%Y")
    date_y = date.strftime("%Y")
    
    with open(f'{name.upper()}_{date_y}/{str(name).lower()}_{date_dmy}.txt', 'a') as file:
        
        write_at_end(file)    
```

```python
def write_at_end(file):
    
#    date = datetime.now()
    file.seek(0, os.SEEK_END)
    date_mdyt = date.strftime('%a %d %b %Y, %I:%M%p')
    file.write(f'{date_mdyt}\n')
```

```python
def makedir_y(name):

#    date = datetime.now()
    date_y = date.strftime("%Y")
    dir_path = Path(f'{name.upper()}_{date_y}')

    try:
        dir_path.mkdir(parents=True, exist_ok=True)

    except Exception as e:
        print(f'Error : {e}')
        
```

```python
def create_sh(name):
    with open(f'openvim.sh', 'w') as file:
        file.write(f'nvim {str(name)}_{date_dmy}.txt')
```
