```python
from datetime import date
from datetime import datetime
from pathlib import Path
import os
import subprocess
```

```python
class CreateDir:

# want to add functionality where if i want multiple directories, program will 
# make all of them recursively

    def __init__(self, name):
        self.name = name
        self.date = datetime.now()
        self.year = self.date.strftime("%Y")

    def makedir(self):

        dir_path = Path(f'{self.name.upper()}_{self.year}')

        try:
            dir_path.mkdir(parents=True, exist_ok=True)

        except Exception as e:
            print(f'Error : {e}')
        
```

```python
class CreateFile:



    def __init__(self, name, typ, location=''):

        self.type = typ
        self.location = location
        self.name = name
        self.date = datetime.now()
        self.date_dmy = self.date.strftime("%d%b%Y")
        self.date_y = self.date.strftime("%Y")
        self.date_mdyt = self.date.strftime('%a %d %b %Y, %I:%M%p')
    

#    def write_at_end(self):
#        file.seek(0, os.SEEK_END)
#        file.write(f'{self.date_mdyt}\n')

    
    def makefile(self):
        
        with open(f'{self.name.upper()}_{self.date_y}/{str(self.name).lower()}_{self.date_dmy}.{self.type}', 'a') as file:

            # appends date at the end
            file.seek(0, os.SEEK_END)
            file.write(f'{self.date_mdyt}\n')
```

```python
filename = 'create'
filetype = 'txt'
createfile = CreateFile(filename,filetype)
createdir = CreateDir(filename)
```

```python
createdir.makedir()

createfile.makefile()
```
